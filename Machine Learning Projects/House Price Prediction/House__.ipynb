```python
x=df[["bedrooms","bathrooms","sqft_living","floors","waterfront","view","grade","sqft_living15","sqft_above","lat","sqft_basement"]]
y=df['price']
```

```python
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=0,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
xtrain=scaler.fit_transform(xtrain)
xtest=scaler.transform(xtest)
```

```python
model=LinearRegression().fit(xtrain,ytrain)
ypred=model.predict(xtest)
accuracy=r2_score(ytest,ypred)
print(accuracy)
```

```python
def eval (xtrain,xtest,ytrain,ytest):
    ypred=model.predict(xtest)
    accuracy=r2_score(ytest,ypred)
    print("accuracy of test",accuracy)
    print("-------------------------------")
    ypredtrain=model.predict(xtrain)
    accuracy=r2_score(ytrain,ypredtrain)
    print("accuracy of train",accuracy)
 # to detect the overfit
```

```python
ob=eval(xtrain,xtest,ytrain,ytest)
print(ob)
# no overfit but low accuracy
```
