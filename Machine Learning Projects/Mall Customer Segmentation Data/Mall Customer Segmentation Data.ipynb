```python
from sklearn.cluster import KMeans
listt=[]
# listt : sum of the variance between the observations in each cluster
for i in range(1,11):
  model=KMeans(n_clusters=i,init='k-means++')
  model.fit(x)
  listt.append(model.inertia_)
fig = plt.figure(figsize = (7,7))
plt.plot(range(1,11),listt, linewidth=4, markersize=12,marker='o',color = 'black')
plt.xticks(np.arange(11))
plt.xlabel("Number of clusters")
plt.ylabel("listt")
plt.show()
```

# Now I'm going to include 3 independant variables such as age, anual income and the spending score in segmentation algorithm

```python
x = df[['Age','Annual Income (k$)','Spending Score (1-100)']].values
```

```python
WCSS = []
for i in range(1,11):
    model = KMeans(n_clusters = i,init = 'k-means++')
    model.fit(x)
    WCSS.append(model.inertia_)

fig = plt.figure(figsize = (7,7))
plt.plot(range(1,11),WCSS, linewidth=4, markersize=12,marker='o',color = 'red')
plt.xticks(np.arange(11))
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
model = KMeans(n_clusters = 5, init = "k-means++", max_iter = 300, n_init = 10, random_state = 0)
predC = model.fit_predict(x)
```
