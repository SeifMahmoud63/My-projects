```python
df['gender'].replace({'Male':0,'Female':1,'Other':3},inplace=True)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df.isnull().sum()
```

```python
df.dropna(axis=1,inplace=True)
```

```python
x=df[['study_hours_per_day','social_media_hours','netflix_hours','sleep_hours','exercise_frequency','mental_health_rating','gender']].values
y=df['exam_score'].values
```

```python
from sklearn.preprocessing import StandardScaler
xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=0,test_size=0.2)
scale=StandardScaler()
xt=scale.fit_transform(xtrain)
xtt=scale.transform(xtest)
```

```python
from sklearn .metrics import r2_score
model=LinearRegression()
model.fit(xt,ytrain)
ypred=model.predict(xtt)
accuracy=r2_score(ytest,ypred)
accuracy
```

```python
r2Train=model.predict(xt)
acc=r2_score(ytrain,r2Train)
acc
# no over fit
```
